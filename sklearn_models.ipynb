```python
#all models in sklearn
import sklearn
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import time
X = np.random.rand(1000000, 5)
y = np.random.rand(1000000, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Linear Regression, Ridge, Lasso, Elastic Net, tune hyperparameters
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.1, 1, 10, 100]}
grid = GridSearchCV(Ridge(), param_grid, cv=5)
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.score(X_test, y_test))

# Logistic Regression, tune hyperparameters
from sklearn.linear_model import LogisticRegression
param_grid = {'C': [0.1, 1, 10, 100]}
grid = GridSearchCV(LogisticRegression(), param_grid, cv=5)

grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.score(X_test, y_test))

# KNN, tune hyperparameters
from sklearn.neighbors import KNeighborsRegressor
param_grid = {'n_neighbors': [1, 3, 5, 7, 9]}
grid = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5)
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.score(X_test, y_test))

# Tree, tune hyperparameters
from sklearn.tree import DecisionTreeRegressor
param_grid = {'max_depth': [1, 3, 5, 7, 9]}
grid = GridSearchCV(DecisionTreeRegressor(), param_grid, cv=5)
grid.fit(X_train, y_train)
print(grid.best_params_)
print(grid.score(X_test, y_test))



```

```python
#dimension reduction
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.manifold import TSNE
from sklearn.manifold import Isomap
from sklearn.manifold import LocallyLinearEmbedding
from sklearn.manifold import MDS
from sklearn.manifold import SpectralEmbedding

X = np.random.rand(100000, 10)
X = np.concatenate((X, X + 1, X - 1), axis=1)
print(X.shape)
#transform and plot

t = time.time()
pca = PCA(n_components=2)
pca.fit_transform(X)
plt.scatter(pca.components_[0], pca.components_[1])
plt.title('PCA')
plt.show()
print('PCA time: ', time.time() - t)

t = time.time()
svd = TruncatedSVD(n_components=2)
svd.fit_transform(X)
plt.scatter(svd.components_[0], svd.components_[1])
plt.title('SVD')
plt.show()
print('SVD time: ', time.time() - t)
```

```python

t = time.time()
tsne = TSNE(n_components=2, perplexity=30, method='barnes_hut', random_state=0)
tsne.fit_transform(X)
plt.scatter(tsne.embedding_[0], tsne.embedding_[1])
plt.title('TSNE')
plt.show()
print('TSNE time: ', time.time() - t)

t = time.time()
isomap = Isomap(n_components=2)
isomap.fit_transform(X)
plt.scatter(isomap.embedding_[:, 0], isomap.embedding_[:, 1])
plt.title('Isomap')
plt.show()
print('Isomap time: ', time.time() - t)

t = time.time()
lle = LocallyLinearEmbedding(n_components=2)
lle.fit_transform(X)
plt.scatter(lle.embedding_[:, 0], lle.embedding_[:, 1])
plt.title('LLE')
plt.show()
print('LLE time: ', time.time() - t)
```

```python

t = time.time()
mds = MDS(n_components=2)
mds.fit_transform(X)
plt.scatter(mds.embedding_[:, 0], mds.embedding_[:, 1])
plt.title('MDS')
plt.show()
print('MDS time: ', time.time() - t)
```

```python

t = time.time()
print('SpectralEmbedding')
se = SpectralEmbedding(n_components=2)
se.fit_transform(X)
plt.scatter(se.embedding_[:, 0], se.embedding_[:, 1])
plt.title('SpectralEmbedding')
plt.show()
print('SpectralEmbedding time: ', time.time() - t)

```

```python
from sklearn.random_projection import GaussianRandomProjection
from sklearn.random_projection import johnson_lindenstrauss_min_dim

# Gaussian Random Projection
grp = GaussianRandomProjection(n_components=2, eps=0.1, random_state=42)
grp.fit_transform(X)
plt.scatter(grp.components_[0], grp.components_[1], alpha=0.5)
plt.title('Gaussian Random Projection')
plt.show()
```
